### Importación de datos



```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda.head()
```

# 3. Calificación promedio de la tienda


```python
calificacion_promedio_tienda1 = tienda['Calificación'].mean()
calificacion_promedio_tienda2 = tienda2['Calificación'].mean()
calificacion_promedio_tienda3 = tienda3['Calificación'].mean()
calificacion_promedio_tienda4 = tienda4['Calificación'].mean()

print(calificacion_promedio_tienda1)
print(calificacion_promedio_tienda2)
print(calificacion_promedio_tienda3)
print(calificacion_promedio_tienda4)
```

```python
calificacion_por_producto_tienda1 = tienda.groupby('Producto')['Calificación'].mean()
calificacion_por_producto_tienda1 = calificacion_por_producto_tienda1.sort_values(ascending=False)

calificacion_por_producto_tienda2 = tienda2.groupby('Producto')['Calificación'].mean()
calificacion_por_producto_tienda2 = calificacion_por_producto_tienda2.sort_values(ascending=False)

calificacion_por_producto_tienda3 = tienda3.groupby('Producto')['Calificación'].mean()
calificacion_por_producto_tienda3 = calificacion_por_producto_tienda3.sort_values(ascending=False)

calificacion_por_producto_tienda4 = tienda4.groupby('Producto')['Calificación'].mean()
calificacion_por_producto_tienda4 = calificacion_por_producto_tienda4.sort_values(ascending=False)

print("Calificación promedio por producto en Tienda 1:")
display(calificacion_por_producto_tienda1)

print("\nCalificación promedio por producto en Tienda 2:")
display(calificacion_por_producto_tienda2)

print("\nCalificación promedio por producto en Tienda 3:")
display(calificacion_por_producto_tienda3)

print("\nCalificación promedio por producto en Tienda 4:")
display(calificacion_por_producto_tienda4)
```

```python
calificacion_promedio_total_por_producto = (calificacion_por_producto_tienda1 + calificacion_por_producto_tienda2 + calificacion_por_producto_tienda3 + calificacion_por_producto_tienda4) / 4
calificacion_promedio_total_por_producto = calificacion_promedio_total_por_producto.sort_values(ascending=False)
print("Calificación promedio total por producto:")
display(calificacion_promedio_total_por_producto)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

tiendas = ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4']
ingresos = [total_ventas_tienda1, total_ventas_tienda2, total_ventas_tienda3, total_ventas_tienda4]
costos_envio_promedio = [costo_envio_promedio_tienda1, costo_envio_promedio_tienda2, costo_envio_promedio_tienda3, costo_envio_promedio_tienda4]
calificacion_promedio = [calificacion_promedio_tienda1, calificacion_promedio_tienda2, calificacion_promedio_tienda3, calificacion_promedio_tienda4]

fig, ax1 = plt.subplots(figsize=(10, 6))

# 1 bar chart for total revenue
sns.barplot(x=tiendas, y=ingresos, ax=ax1, palette='viridis', hue=tiendas, legend=False)
ax1.set_xlabel('Tienda')
ax1.set_ylabel('Ingresos Totales', color='teal')
ax1.tick_params(axis='y', labelcolor='teal')
ax1.set_title('Ingresos Totales, Costo de Envío Promedio y Calificación Promedio por Tienda')

# 2 y-axis for average shipping cost
ax2 = ax1.twinx()
ax2.plot(tiendas, costos_envio_promedio, color='orangered', marker='o', linewidth=2, label='Costo de Envío Promedio')
ax2.set_ylabel('Costo de Envío Promedio', color='orangered')
ax2.tick_params(axis='y', labelcolor='orangered')

# 3 y-axis for average rating
ax3 = ax1.twinx()
ax3.spines['right'].set_position(('outward', 60))
ax3.plot(tiendas, calificacion_promedio, color='purple', marker='x', linewidth=2, label='Calificación Promedio')
ax3.set_ylabel('Calificación Promedio', color='purple')
ax3.tick_params(axis='y', labelcolor='purple')
ax3.set_ylim(0, 5) # 0 and 5

fig.legend(loc="upper left", bbox_to_anchor=(0.1, 0.9))

plt.tight_layout()
plt.show()
```
